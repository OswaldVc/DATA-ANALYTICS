````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python

# Creamos un DataFrame simple de prueba
data = {
    'Edad': [23, 45, 31, 22, 41],
    'Ingresos': [35000, 65000, 48000, 30000, 52000]
}

df = pd.DataFrame(data)
df
````

## 📈 Visualización rápida


````python
sns.scatterplot(data=df, x='Edad', y='Ingresos')
plt.title('Relación entre Edad e Ingresos')
plt.show()
````
